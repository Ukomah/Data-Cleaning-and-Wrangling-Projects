**Task 7** Write a function to clean and convert curency values, then apply the function on `"value", "wage", "release_clause"`

```python
# Function to clean and convert currency values
def convert_currency(value):
    """Convert FIFA currency strings (e.g., '€110M', '€500K') to numeric values."""
    
    if isinstance(value, str):  # Check if value is a string
        value = value.replace("€", "")  # Remove the euro sign
        if "M" in value:  
            value = value.replace("M", "")  # Remove "M"
            return float(value) * 1_000_000  # Convert to millions
        elif "K" in value:  
            value = value.replace("K", "")  # Remove "K"
            return float(value) * 1_000  # Convert to thousands
        return float(value)  # Convert plain numbers to float
    return value  # Return unchanged if already a number



df[["value", "wage", "release_clause"]] = df[["value", "wage", "release_clause"]].map(convert_currency)
```

**Task 8:** Write a function that will clean and convert the observations in `height` and `weight` into numericals.


```python
def convert_height(height):
    """Convert height from feet'inches format to centimeters."""
    if isinstance(height, str):
        if "cm" in height:
            height = height.replace("cm", "")
            return float(height)
        elif "'" in height:  # Ensure it's a string and contains feet notation
            feet, inches = height.split("'")  # Split at apostrophe
            height = (int(feet) * 30.48) + (int(inches.replace('"', "")) * 2.54)  # Convert to cm
            return float(height)
    return height
    

df["height"] = df["height"].apply(convert_height)
```

```python
def convert_weight(weight):
    """Convert weight from lbs to kg."""
    if isinstance(weight, str):
        if "kg" in weight:
            weight = weight.replace("kg", "")
            return float(weight)
        elif "lbs" in weight:  # Ensure it's a string with "lbs"
            weight = round(int(weight.replace("lbs", "")) * 0.453592, 1)  # Convert to kg and round
            return weight
    return weight

# Assuming df is your DataFrame
df["weight"] = df["weight"].apply(convert_weight)
```

**Task 9:** Rename `height` and `weight` to include their unit of measurement.

```python
df.rename(columns={"height": "height(cm)", "weight":"weight(kg)"}, inplace=True)
```

**Task 10:** Columns w/f, sm, ir contain special characters (★) → Convert them to integers.

```python
df[["w/f", "sm", "ir"]] = df[["w/f", "sm", "ir"]].replace("★", "", regex=True).astype(int)
```

**Task 11:**  Calculate the difference between contract years and replace the existing values with the duration.

```python
# Function to extract contract duration
def extract_duration(contract):
    try:
        # Check if contract follows the "YYYY ~ YYYY" format
        if " ~ " in contract:
            start_year, end_year = contract.split(" ~ ")
            return int(end_year) - int(start_year)  # Compute duration
        else:
            return "On Loan"  # Handle cases like "On Loan"
    except:
        return None  # Catch any unexpected formats
    

# Apply the function to the contract column
df["contract_duration"] = df["contract"].apply(extract_duration)

df.drop(columns=["contract"], inplace=True)
```

**Task 12:** Convert the `joined` Column to datetime

```python
# Convert the 'joined' column to datetime format
df["joined"] = pd.to_datetime(df["joined"], format="%b %d, %Y")
```
